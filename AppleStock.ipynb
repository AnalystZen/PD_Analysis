```python
from matplotlib import dates
import matplotlib as plt
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
raw = yf.download('AAPL', start='2010-01-01', end='2019-12-31')
```

```python
# mapping out datas frame, refigure size is in inches
(raw
 .iloc[:,[1,2]]
 .plot(figsize= (5,2))
)
```

```python
# mapping out datas frame, remapping data based on time
(raw
 .iloc[:,[1,2]]
 .resample("ME")
 .mean()
)
```

```python
# Plotting new chart based on month
(raw
 .iloc[:,[1,2]]
 .resample("ME")
 .mean()
 .plot(figsize= (5,2))
)
```

```python
# taking the dataframe and resampling by year
(raw
 .iloc[:,[1,2]]
 .resample("YE")
 .mean()
 .plot(figsize= (5,2))
)
```

```python
# Apple volume by month/year
(raw
 .iloc[:,[4]]
 .resample("ME")
 .mean()
 .plot(figsize= (5,2))
)
```

```python
# Apple volume by month/year
(raw
 .iloc[:,[0,3]]
 .resample("YE")
 .mean()
 .plot(figsize= (5,2))
)
```

```python
(raw
 .iloc[:,[0,3]]
 .resample("YE")
 .mean()
 .plot(figsize=(5,2))
)
```

```python
# Calculate percent change over time
(raw
 .Close
 .pct_change()
 .add(1)
 .cumprod()
 .sub(1)
 .plot(figsize=(5,2))
)
```
